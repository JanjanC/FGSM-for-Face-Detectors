```python
#THIS IS THE SCRIPT FOR RENAMING FILES
import os
import glob
from pathlib import Path

SOURCE_PATH = os.path.join(os.getcwd())
for x in os.walk(SOURCE_PATH):
    if x[0].endswith('csv'):
        print('Folder: ', x[0])
        for file in glob.glob(os.path.join(x[0], '*.csv')):
            print('Old Name: ', file)
    
            filename = os.path.basename(file) #extract file name        
            parent = x[0].split('\\')[-2] #get folder name
            
            destination =  os.path.join(x[0], parent + '_' + filename) # destination + file name
            print('New Name: ', destination)
            
            os.rename(file, destination) #rename files
```
