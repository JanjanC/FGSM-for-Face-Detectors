```python
from scipy import statsfrom scipy import stats
import sklearn
df = df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

```python
df.describe()
```

```python
# Log Transform
df['w'] = np.log(df['w'])
df['h'] = np.log(df['h'])
df.describe()
```

```python
# Standardize h, w
for column in df:
    if column != 'sigma':
#         print(column)
        df[column] = (df[column] - df[column].mean()) / df[column].std()
#         df[column] = sklearn.preprocessing.Normalizer().fit_transform(df[[column]])

df.describe()
```

```python
def train_test_val_split(df):
  X_train = df.sample(frac=0.8, random_state=0)
  X_test = df.drop(X_train.index)
    
#   X_test_sample = X_test.sample(frac=0.5, random_state=10 * i + j)
#   X_val = X_test.drop(X_test_sample.index)

  y_train = X_train['sigma'].to_numpy().astype(np.float64)
  y_test = X_test['sigma'].to_numpy().astype(np.float64)
#   y_val = X_val['sigma'].to_numpy().astype(np.float64)

  X_train = X_train.drop(['area', 'sigma', 'x', 'y'], axis=1).to_numpy().astype(np.float64)
  X_test = X_test.drop(['area', 'sigma',  'x', 'y'], axis=1).to_numpy().astype(np.float64)

#   X_train = X_train[['w', 'h']]
#   X_test = X_test[['w', 'h']]

#   X_val = X_val.drop(['area', 'sigma', 'x', 'y', 'destination', 'frame'], axis=1).to_numpy().astype(np.float64)
#   X_train = X_train['w', 'h'].to_numpy().astype(np.float64)
#   X_test_sample = X_test_sample['w', 'h'].to_numpy().astype(np.float64)
#   X_val = X_val['w', 'h'].to_numpy().astype(np.float64)

  # print(X_train.shape)
  # print(X_test_sample.shape)
  # print(X_val.shape)

  # print(y_train.shape)
  # print(y_test.shape)
  # print(y_val.shape)

  return X_train, X_test, y_train, y_test
  
```

```python
X_train, X_test, y_train, y_test = train_test_val_split(df)
```
